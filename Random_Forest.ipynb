```python
#Import
import numpy as np
import pandas as pd
import warnings
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

warnings.simplefilter("ignore")
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Data Science - Equipo: Zelada, Caceres M., Rodriguez S./Segunda Pre-Entrega/combats.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Crehana/Data Science/Machine Learning/Random Forest/Copia de pokemon_csv.csv')
```

```python
df_2 = df2.copy(deep=True)
df_2.drop('nombre', inplace = True, axis = 1)
df_2.drop(['tipo1','tipo2'], inplace = True, axis = 1)
df_2.drop(['Unnamed: 0', 'Unnamed: 0.5', 'Unnamed: 0.4', 'Unnamed: 0.3',	'Unnamed: 0.2', 'Unnamed: 0.1'], inplace=True, axis = 1)
df_2.head()
```

```python
difaltura_m = []
difpeso_kg = []
difhp = []
difataque = []
difdefensa = []
difataque_esp = []
difdefensa_esp = []
difvelocidad = []

for i in range(len(df1.Winner)):

  difaltura_m.append(df_2.altura_m[df1.First_pokemon[i]] - df_2.altura_m[df1.Second_pokemon[i]])

  difpeso_kg.append(df_2.peso_kg[df1.First_pokemon[i]] - df_2.peso_kg[df1.Second_pokemon[i]])

  difhp.append(df_2.hp[df1.First_pokemon[i]] - df_2.hp[df1.Second_pokemon[i]])

  difataque.append(df_2.ataque[df1.First_pokemon[i]] - df_2.ataque[df1.Second_pokemon[i]])

  difdefensa.append(df_2.defensa[df1.First_pokemon[i]] - df_2.defensa[df1.Second_pokemon[i]])

  difataque_esp.append(df_2.ataque_esp[df1.First_pokemon[i]] - df_2.ataque_esp[df1.Second_pokemon[i]])

  difdefensa_esp.append(df_2.defensa_esp[df1.First_pokemon[i]] - df_2.defensa_esp[df1.Second_pokemon[i]])

  difvelocidad.append(df_2.velocidad[df1.First_pokemon[i]] - df_2.velocidad[df1.Second_pokemon[i]])


df1['difaltura_m'] = difaltura_m
df1['difpeso_kg'] = difpeso_kg
df1['difhp'] = difhp
df1['difataque'] = difataque
df1['difdefensa'] = difdefensa
df1['difataque_esp'] = difataque_esp
df1['difdefensa_esp'] = difdefensa_esp
df1['difvelocidad'] = difvelocidad


binwin = []

for i in range(len(df1.Winner)):
  if df1.Winner[i] == df1.First_pokemon[i]:
    binwin.append(0)
  elif df1.Winner[i] == df1.Second_pokemon[i]:
    binwin.append(1)
  else:
    binwin.append(np.nan)

df1['binwin'] = binwin

df1.head()
```

## Train-Test Split

```python
y = df1.binwin
combat_features = ['First_pokemon', 'Second_pokemon', 'difhp', 'difataque', 'difdefensa', 'difataque_esp', 'difdefensa_esp', 'difvelocidad']
X = df1[combat_features]

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=33,
                                                    stratify= y)


print('Tamaño del set de entrenamiento:', X_train.shape)
print('Tamaño del set de testeo:', X_test.shape)
```

## Random Forest Classifier

```python
forest_model_cl = RandomForestClassifier(random_state=1)
forest_model_cl.fit(X_train, y_train)
```

```python
y_pred = forest_model_cl.predict(X_test)
```

```python
print(y_pred)
```

```python
print('Matriz de confusion:')
print(metrics.confusion_matrix(y_test, y_pred))
print('Classification report:')
print(metrics.classification_report(y_test,y_pred))
print('-----------------------------------------------------------------------------')
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print('Precision:',metrics.precision_score(y_test, y_pred))
print('Recall:',metrics.recall_score(y_test, y_pred))
print('F1:',metrics.f1_score(y_test, y_pred))
```

## Random Forest Regressor

```python
forest_model_reg = RandomForestRegressor(random_state=1)
forest_model_reg.fit(X_train, y_train)
```

```python
predictions = forest_model_reg.predict(X_test)
```

```python
print(predictions)
```

```python
predictionsint = []

for i in range(len(predictions)):
  if int(predictions[i]) - X_test.First_pokemon[X_test.index[i]] > int(predictions[i]) - X_test.Second_pokemon[X_test.index[i]]:
    predictionsint.append(X_test.First_pokemon[X_test.index[i]])
  else:
    predictionsint.append(X_test.Second_pokemon[X_test.index[i]])

print(predictionsint)
```

```python
results = X_test.copy(deep=True)
results.drop(['difhp', 'difataque', 'difdefensa', 'difataque_esp', 'difdefensa_esp',
 'difvelocidad'], inplace = True, axis = 1)
results['prediction'] = predictionsint
results.sample(10)
```

Dado que la predicción generada no es binaria, sino discreta, deberemos transformarlas. Para ello estableceremos un limite:

En este caso puntual utilizaremos 0.5 como limite. 
En el rango 0.5 - 1 consideraremos la respuesta como 1/True
en el rango 0 - 0.49 consideraremos la predicción como 0/False

**Sugerencia:** Probar con diferentes rangos y comparar resultados

```python
y_pred = []

for i in predictions:
  if i >= 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

print(np.array(y_pred))
```

```python
print('Matriz de confusion:')
print(metrics.confusion_matrix(y_test, y_pred))
print('Classification report:')
print(metrics.classification_report(y_test,y_pred))
print('-----------------------------------------------------------------------------')
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print('Precision:',metrics.precision_score(y_test, y_pred))
print('Recall:',metrics.recall_score(y_test, y_pred))
print('F1:',metrics.f1_score(y_test, y_pred))
```
